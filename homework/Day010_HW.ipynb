## requests + BeautifulSoup


```python
import requests 
from bs4 import BeautifulSoup
url = 'https://www.dcard.tw/f'
r = requests.get(url)
r.encoding = 'utf-8'
response = r.text[0:5000]
soup = BeautifulSoup(response)

print(soup.title.text)
print('-------------------')
print(soup)
```

## Grab + PyQuery

```python
from grab import Grab
g = Grab()
resp = g.go('https://www.dcard.tw/f')
resp.body[0:5000]
print(type(resp))
```

```python
from pyquery import PyQuery as pq
doc = pq(resp.body)
print(type(doc))

title = doc('title')
print(type(title))
print(title.text())
print('----------------------')
print(doc)
```
