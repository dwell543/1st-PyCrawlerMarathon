## 作業目標

* 請利用 API: https://www.dcard.tw/_api/forums/pet/posts?popular=true 回答下列問題：

1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

```python
import requests
r = requests.get("https://www.dcard.tw/_api/forums/pet/posts?popular=true")
r.text
```

```python
import json
data = json.loads(r.text)
for d in data:
    print(d['title'])
```

```python
# 1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
print('一次會回傳的資料:',len(data),'筆')

#每一筆資料包含哪些欄位？
print('每筆資料欄位:',data[0].keys())
```

```python
# 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」

for d in data:
    print('標題:',d['title'],'貼文時間:',d['createdAt'],'留言人數:',d['commentCount'],'按讚人數:',d['likeCount'])
```

```python
# 3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」
import requests
from statistics import mean
r_pop = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
data_pop = json.loads(r_pop.text)
m_pop_comment = mean([pop_c_m['commentCount'] for pop_c_m in data_pop])
m_pop_like = mean([pop_l_m['likeCount'] for pop_l_m in data_pop])

r_npop = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false')
data_npop = json.loads(r_npop.text)
m_npop_comment = mean([npop_c_m['commentCount'] for npop_c_m in data_npop])
m_npop_like = mean(npop_l_m['likeCount'] for npop_l_m in data_npop)

print("熱門文章平均留言人數:",m_pop_comment)
print("熱門文章平均按讚人數:",m_pop_like)
print("非熱門文章平均留言人數:",m_npop_comment)
print("非熱門文章平均按讚人數:",m_npop_like)
```
