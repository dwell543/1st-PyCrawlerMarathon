```python
import json
data = json.loads(r.text)
for d in data:
    print(d['title'])
```

```python
# 2. å–å‡ºæ¯ä¸€ç­†è³‡æ–™çš„ã€Œæ¨™é¡Œã€ã€ã€Œè²¼æ–‡æ™‚é–“ã€ã€ã€Œç•™è¨€äººæ•¸ã€ã€ã€ŒæŒ‰è®šäººæ•¸ã€

for d in data:
    print('æ¨™é¡Œ:',d['title'],'è²¼æ–‡æ™‚é–“:',d['createdAt'],'ç•™è¨€äººæ•¸:',d['commentCount'],'æŒ‰è®šäººæ•¸:',d['likeCount'])
```

```python
# 3. è¨ˆç®—ç†±é–€/éç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€èˆ‡ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€
import requests
from statistics import mean
r_pop = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
data_pop = json.loads(r_pop.text)
m_pop_comment = mean([pop_c_m['commentCount'] for pop_c_m in data_pop])
m_pop_like = mean([pop_l_m['likeCount'] for pop_l_m in data_pop])

r_npop = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false')
data_npop = json.loads(r_npop.text)
m_npop_comment = mean([npop_c_m['commentCount'] for npop_c_m in data_npop])
m_npop_like = mean(npop_l_m['likeCount'] for npop_l_m in data_npop)

print("ç†±é–€æ–‡ç« å¹³å‡ç•™è¨€äººæ•¸:",m_pop_comment)
print("ç†±é–€æ–‡ç« å¹³å‡æŒ‰è®šäººæ•¸:",m_pop_like)
print("éç†±é–€æ–‡ç« å¹³å‡ç•™è¨€äººæ•¸:",m_npop_comment)
print("éç†±é–€æ–‡ç« å¹³å‡æŒ‰è®šäººæ•¸:",m_npop_like)
```
