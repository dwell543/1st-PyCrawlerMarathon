## 1. 取出今天所有的新聞

```python
#靜態取出所有新聞

import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, 'lxml')

for d in soup.find(class_='part_list_2').find_all('h3'):
    print(d.find(class_='date').text, d.find('a').get_text())
```

```python
#利用selenium 抓取利用JavaScript動態載入的新聞
from selenium import webdriver
from bs4 import BeautifulSoup

browser = webdriver.Chrome(executable_path='chromedriver')
browser.get('https://www.ettoday.net/news/news-list.htm')
```

# selenium 自動化讀取JS資料
#https://medium.com/@yanweiliu/python%E7%88%AC%E8%9F%B2%E5%AD%B8%E7%BF%92%E7%AD%86%E8%A8%98-%E4%BA%8C-selenium%E8%87%AA%E5%8B%95%E5%8C%96-ab0a27a94ff2

```python
#每個2秒自動下滑
import time
for i in range(10):
    # 延遲2秒後...
    time.sleep(2)
    ##例如使用JS移動到網頁最下方：driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    #　　window.scrollTo(0,1000); // 垂直滾動到1000的位置
    browser.execute_script("window.scrollTo(0, 200000);")
```

```python
#取得資料丟到BeautifulSoup分析
#網頁原始碼
html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")

for d in soup.find(class_="part_list_2").find_all("h3"):
    print(d.find(class_="date").text, d.find_all("a")[-1].text)
```

## 2. 取出現在時間兩小時內的新聞

```python
# 限制2個小時的時間
from datetime import datetime, timedelta
twoHours_ago = datetime.now() - timedelta(hours=2)

#取出資料
news = []

for d in soup.find(class_='part_list_2').find_all('h3'):
    #超過2小時就結束
    if datetime.strptime(d.find(class_='date').text, '%Y/%m/%d %H:%M') < twoHours_ago:
        break
    news.append({
        '時間':d.find(class_='date').text,
        '標題':d.find_all('a')[-1].text
    })
    
import pandas as pd
pd.DataFrame(news)
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

#timedelta https://python3-cookbook.readthedocs.io/zh_CN/latest/c03/p12_convert_days_to_seconds_and_others.html

```python
date_time = datetime.now() - timedelta(days=3)
print('取出前三天時間:', date_time, type(date_time))

#time.strftime() 函數可以依照指定的格式將 struct_time 時間資料轉換為文字輸出
#time.strptime() 函數則是跟 time.strftime() 函數相反，它是依據指定的格式，解析文字資料中的時間，輸出 struct_time 格式的時間

date_time_year = date_time.strftime("%Y")
date_time_month = date_time.strftime("%m")
date_time_day = date_time.strftime("%d")

print("前三天月&日:",date_time_month,"&",date_time_day)

time_3hago = datetime.strptime(date_time.strftime("%Y/%m/%d") + ' 15', '%Y/%m/%d %H')
time_5hago = datetime.strptime(date_time.strftime("%Y/%m/%d") + ' 17', '%Y/%m/%d %H')

print(time_3hago, type(time_3hago))
print(time_5hago, type(time_5hago))
```

```python
#selenium操作 https://huilansame.github.io/huilansame.github.io/archivers/drop-down-select
#選擇點選日期
from selenium.webdriver.support.ui import Select
selY = Select(browser.find_element_by_id("selY"))
selY.select_by_value(date_time_year)
selM = Select(browser.find_element_by_id("selM"))
selM.select_by_value(date_time_month)
selD = Select(browser.find_element_by_id("selD"))
selD.select_by_value(date_time_day)
browser.find_element_by_id('button').click()
```

```python
#利用selenium持續執行下滑的動作，直到取得目標時間內的新聞
while True:
    browser.execute_script("window.scrollTo(0, 1000000);")
    time.sleep(1)
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, "html5lib") 
    new_date = soup.find_all(class_="part_list_2")[-1].find_all("h3")[-1].find(class_="date").text
    if datetime.strptime(new_date, "%Y/%m/%d %H:%M") < time_3hago:
        break
```

```python
news_3_5 = []


for d_35 in soup.find(class_='part_list_2').find_all('h3'):
    time_3_5 = datetime.strptime(d_35.find(class_='date').text, '%Y/%m/%d %H:%M')
    if time_3_5 < time_3hago:
        break
    if time_3_5 < time_5hago:
         news_3_5.append({
        '時間': d_35.find(class_='date').text,
        '標題': d_35.find_all('a')[-1].text
        })
        
   
pd.DataFrame(news_3_5)
```
