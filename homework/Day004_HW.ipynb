### 比較一下範例檔案中的「r.text」與「json.loads(r.text)」讀出來的內容有什麼差異

```python
import requests
r = requests.get('https://api.github.com/events')
r.text[0:500]
```

```python
import json
json.loads(r.text)[0:1]
```

```python
import requests,json
r = requests.get('https://api.github.com/events')

print(type(r.text),'and',type(json.loads(r.text)))
```

### 自行尋找一個合適的 API 接口做練習，並且查看其回傳內容

* https://cat-fact.herokuapp.com/facts (來源：https://alexwohlbruck.github.io/cat-facts/)
* http://odata.wra.gov.tw/v4/RealtimeWaterLevel (來源：https://data.gov.tw/dataset/25768)


```python
import requests
#get 請求
r = requests.get('https://api.kcg.gov.tw/api/service/Get/f2c5f476-3f1b-4904-acd4-c360cbfb8773')
#伺服器回應狀態
print("伺服器回應:",r.status_code)
r.text
```

```python
import json
r = requests.get('http://opendata.epa.gov.tw/ws/Data/UV/?format=json')
print("伺服器回應:",r.status_code)
json.loads(r.text)
```

```python
#查詢URL參數
my_params = {'PublishTime':'2019-12-04 14:00','UVI':'3.91'}
r = requests.get('http://opendata.epa.gov.tw/ws/Data/UV/?format=json',params = my_params)
print(r.url)
```
