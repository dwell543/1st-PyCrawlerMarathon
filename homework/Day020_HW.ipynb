```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(r.text, 'html.parser')
table = soup.find('table', class_='TABLE_G')
print(table)
```

```python
items = table.find_all('tr')[1:]
#print(items)
air = {}

for d in items:
    if(len(d.find_all('td'))) == 5:
        air_item =d.find_all('td')[0].get_text()
        air[air_item] = {}
        date_1 = d.find_all('td')[2].get_text()
        air[air_item][date_1] = d.find_all('td')[3].get_text()
        #print(air)
    else:
        date_2 = d.find_all('td')[0].get_text()
        air[air_item][date_2] = d.find_all('td')[1].get_text()
air
```
